### Dependencies

```python
import requests
import re
import pandas as pd
from bs4 import BeautifulSoup
from lxml import html
import datetime
# from selenium import webdriver
# from selenium.common.exceptions import NoSuchElementException
# from selenium.webdriver.common.keys import Keys
```

```python
print(datetime.datetime.now())
driver = webdriver.Safari()
driver.get("https://www.cargurus.com/Cars/inventorylisting/viewDetailsFilterViewInventoryListing.action?sourceContext=cargurus&newSearchFromOverviewPage=true&inventorySearchWidgetType=AUTO&entitySelectingHelper.selectedEntity=&entitySelectingHelper.selectedEntity2=&zip=03824&distance=100&searchChanged=true&modelChanged=true&filtersModified=true&sortType=undefined&sortDirection=undefined")
R = []
while True:
    response = requests.get("https://www.cargurus.com/Cars/inventorylisting/viewDetailsFilterViewInventoryListing.action?sourceContext=cargurus&newSearchFromOverviewPage=true&inventorySearchWidgetType=AUTO&entitySelectingHelper.selectedEntity=&entitySelectingHelper.selectedEntity2=&zip=03824&distance=100&searchChanged=true&modelChanged=true&filtersModified=true&sortType=undefined&sortDirection=undefined")
    try:
        soup = BeautifulSoup(driver.page_source,"lxml")
        Title = soup.findAll(class_ = "cg-dealFinder-result-model")
        for tl in Title:
            R1 = (tl.get_text()).split("\n")
            R.append(R1)
        driver.find_element_by_xpath('//*[contains(concat( " ", @class, " " ), concat( " ", "js-go-to-next-page", " " ))]').click()
    except:
        break

print(datetime.datetime.now())
```

```python
df = pd.DataFrame(R)
df
```

```python
df.to_csv("Carguru.csv")
```
